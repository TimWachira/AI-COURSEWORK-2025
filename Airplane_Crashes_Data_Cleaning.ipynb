**Create fatality_locations dataframe**

```python
fatality_locations = df[['Date', 'Location', 'Aboard', 'Fatalities']].copy()
```

**Find the date of the highest number of fatalities**

```python
fatality_locations.loc[fatality_locations['Fatalities'].idxmax()]
```

**Recorded crashes where fatalities = 0**

```python
no_fatalities = fatality_locations[fatality_locations['Fatalities'] == 0]
len(no_fatalities)
```

**Split column to have 2 columns**

```python
location_split = fatality_locations['Location'].str.split(',', n=1, expand = True)
fatality_locations['Region'] = location_split[0]
fatality_locations['State/Country'] = location_split[1]
```

**Order fatalities from highest to owest and select first 100 fatalities**

```python
top100 = fatality_locations.sort_values(by = 'Fatalities', ascending = False).head(100)
```

**Generate a pie chart of top 25 number of fatalities per country/U.S. State**

```python
import seaborn as sns
import matplotlib.pyplot as plt

top25 = top100.groupby('State/Country')['Fatalities'].sum().nlargest(25)
top25.plot.pie(autopct = '%1.1f%%', figsize = (8,8))
plt.title('Top 25 Number of Fatalities per Country/U.S. State')
plt.show()
```
